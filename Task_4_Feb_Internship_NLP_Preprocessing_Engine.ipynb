
# 🚀 Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.  

"Love" → may represent a proper noun or sentence start  
"love" → general verb/noun  

After lowercasing, both become the same token → improves consistency.

---

## 2. What happens if stopwords are not removed?
If stopwords are not removed:
- Model gets noisy and less meaningful features  
- Increased dimensionality  
- Reduced model performance  
- Example: "the", "is", "and" add little semantic value  

---

## 3. Two cases where removing stopwords is harmful

**Sentiment Analysis**  
"I am not happy" → removing "not" changes meaning completely  

**Question Answering**  
"What is your name?" → removing "what" destroys intent  

---

## 4. Stemming vs Lemmatization

| Feature        | Stemming            | Lemmatization         |
|----------------|---------------------|------------------------|
| Approach       | Rule-based cutting  | Dictionary-based       |
| Output         | May be incorrect    | Meaningful word        |
| Example        | running → run       | running → run          |
| Accuracy       | Low                 | High                   |

```python
import re

# 1. Remove Numbers
def remove_numbers(text):
    return re.sub(r"\d+", "", text)

# 2. Remove Extra Spaces
def remove_extra_spaces(text):
    return re.sub(r"\s+", " ", text).strip()

# 3. Handle Repeated Characters
def handle_repeated_chars(text):
    return re.sub(r'(.)\1+', r'\1', text)

# 4. Convert to Lowercase
def to_lowercase(text):
    return text.lower()

# 5. Remove Short Tokens (≤2 except 'no', 'not')
def remove_short_tokens(text):
    tokens = text.split()
    filtered = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    return " ".join(filtered)

# 6. Remove URLs & Emails
def remove_urls_emails(text):
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    return text

#user input

print("\n--- NLP Preprocessing Interactive Program ---\n")

# 1. Remove Numbers
text1 = input("Enter sentence for NUMBER REMOVAL: ")
print("Output:", remove_numbers(text1))
print("-" * 50)

# 2. Remove Extra Spaces
text2 = input("Enter sentence for EXTRA SPACE REMOVAL: ")
print("Output:", remove_extra_spaces(text2))
print("-" * 50)

# 3. Handle Repeated Characters
text3 = input("Enter sentence for REPEATED CHARACTER HANDLING: ")
print("Output:", handle_repeated_chars(text3))
print("-" * 50)

# 4. Lowercase Conversion
text4 = input("Enter sentence for LOWERCASE CONVERSION: ")
print("Output:", to_lowercase(text4))
print("-" * 50)

# 5. Remove Short Tokens
text5 = input("Enter sentence for SHORT TOKEN REMOVAL: ")
print("Output:", remove_short_tokens(text5))
print("-" * 50)

# 6. Remove URLs & Emails
text6 = input("Enter sentence for URL & EMAIL REMOVAL: ")
print("Output:", remove_urls_emails(text6))
print("-" * 50)
```

```python
# 📊 Task 4: Token Analytics
```

```python
def token_analysis(tokens):
    if not tokens:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_length, 2)


for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg_len)
    print("-" * 50)
```

```python
# 🔄 Task 6: Full Pipeline
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(test_sentences)

print("All Tokens:\n", result["tokens"])
print("\nClean Sentences:\n", result["clean_sentences"])
```
